The planned generators dataset in this survey could be used to identify and track the evolution of proposed generating projecs. This information is publicly available on the EIA's website: [Form-860](https://www.eia.gov/electricity/data/eia860
"EIA Form-860")

Isolating this information on a regular basis can be tedious, time consuming, and if done manually, can result in datasets that contain errors (i.e. typos, etc.).

```python
# Creating a new column for the first dataset.

pg17 = proposed_gen_2017

pg17['Plant_Name_Gen_ID']=pg17.apply(lambda x:'%s_%s' % (x["Plant Name"],x["Generator ID"]),axis=1)

pg17.head()
```

```python
# Setting the new column as index for the first dataset.

pg17.set_index("Plant_Name_Gen_ID", inplace=True)
pg17.head()
```

```python
# Shrinking the first dataset by dropping coloumns to keep only relevant info.

pg17 = pg17.drop(['Unit Code',
       'Ownership', 'Duct Burners',
       'Can Bypass Heat Recovery Steam Generator?',
       'RTO/ISO LMP Node Designation',
       'RTO/ISO Location Designation for Reporting Wholesale Sales Data to FERC',
       'Nameplate Power Factor',
       'Summer Capacity (MW)', 'Winter Capacity (MW)',
       'Associated with Combined Heat and Power System', 'Sector Name',
       'Sector', 'Previously Canceled', 'Energy Source 1', 'Energy Source 2',
       'Energy Source 3', 'Energy Source 4', 'Energy Source 5',
       'Energy Source 6', 'Turbines or Hydrokinetic Buoys',
       'Fluidized Bed Technology?', 'Pulverized Coal Technology?',
       'Stoker Technology?', 'Other Combustion Technology?',
       'Subcritical Technology?', 'Supercritical Technology?',
       'Ultrasupercritical Technology?', 'Solid Fuel Gasification System?',
       'Carbon Capture Technology?', 'Multiple Fuels?',
       'Switch Between Oil and Natural Gas?', 'Cofire Fuels?'], axis="columns")
pg17.head()
```

```python
# Sorting the updated first dataset by the new column.

pg17.sort_values(by=["Plant_Name_Gen_ID"])
pg17.head()
```

```python
# Creating a new column for the second dataset.

pg18 = proposed_gen_2018

pg18['Plant_Name_Gen_ID']=pg18.apply(lambda x:'%s_%s' % (x["Plant Name"],x["Generator ID"]),axis=1)

pg18.head()
```

```python
# Setting the new column as index for the second dataset.

pg18.set_index("Plant_Name_Gen_ID", inplace=True)
pg18.head()
```

```python
# Shrinking the first dataset by dropping coloumns to keep only relevant info.

pg18 = pg18.drop(['Unit Code',
       'Ownership', 'Duct Burners',
       'Can Bypass Heat Recovery Steam Generator?',
       'RTO/ISO LMP Node Designation',
       'RTO/ISO Location Designation for Reporting Wholesale Sales Data to FERC',
       'Nameplate Power Factor',
       'Summer Capacity (MW)', 'Winter Capacity (MW)',
       'Associated with Combined Heat and Power System', 'Sector Name',
       'Sector', 'Previously Canceled', 'Energy Source 1', 'Energy Source 2',
       'Energy Source 3', 'Energy Source 4', 'Energy Source 5',
       'Energy Source 6', 'Turbines or Hydrokinetic Buoys',
       'Fluidized Bed Technology?', 'Pulverized Coal Technology?',
       'Stoker Technology?', 'Other Combustion Technology?',
       'Subcritical Technology?', 'Supercritical Technology?',
       'Ultrasupercritical Technology?', 'Solid Fuel Gasification System?',
       'Carbon Capture Technology?', 'Multiple Fuels?',
       'Switch Between Oil and Natural Gas?', 'Cofire Fuels?'], axis="columns")
pg18.head()
```

```python
# Sorting the updated second dataset by the new column.

pg18.sort_values(by=["Plant_Name_Gen_ID"])
pg18.head()
```

```python
# Isolating the differences in the indexes

idx1 = pg17.index
idx2 = pg18.index
idx1.difference(idx2)
```

```python
# Creating a dataset comprised of the diffrences between the two original datasets.

pg1718_diff = pg18.loc[idx2.difference(idx1), :]
pg1718_diff.head()
```

```python
# Inspecting updated dataset

pg1718_diff.describe(include=object)
```

```python
# Inspecting data types

pg1718_diff.dtypes
```

```python
# Rename Nameplate Capacity column

pg1718_diff = pg1718_diff.rename(columns={"Nameplate Capacity (MW)": "Nameplate Capacity"})
pg1718_diff.head()
```

```python
# Changing Nameplate Capacity type to float

pg1718_diff["Nameplate Capacity"] = pg1718_diff["Nameplate Capacity"].astype(float)
pg1718_diff.dtypes
```
