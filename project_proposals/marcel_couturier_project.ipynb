```python
# Creating a new column for the Prior_Year dataset.

prior_yr = proposed_gen_prior_yr
prior_yr['Plant_Name_Gen_ID']=prior_yr.apply(
    lambda x:'%s_%s' % (x["Plant Name"],x["Generator ID"]),axis=1)
prior_yr.head(3)
```

```python
# Setting the new column as index for the Prior_Year dataset.

prior_yr.set_index("Plant_Name_Gen_ID", inplace=True)
prior_yr.head(3)
```

```python
# Shrinking the Prior_Year dataset by dropping coloumns to keep only relevant info.

prior_yr = prior_yr.drop(['Unit Code',
       'Ownership', 'Duct Burners',
       'Can Bypass Heat Recovery Steam Generator?',
       'RTO/ISO LMP Node Designation',
       'RTO/ISO Location Designation for Reporting Wholesale Sales Data to FERC',
       'Nameplate Power Factor',
       'Summer Capacity (MW)', 'Winter Capacity (MW)',
       'Associated with Combined Heat and Power System', 'Sector Name',
       'Sector', 'Previously Canceled', 'Energy Source 1', 'Energy Source 2',
       'Energy Source 3', 'Energy Source 4', 'Energy Source 5',
       'Energy Source 6', 'Turbines or Hydrokinetic Buoys',
       'Fluidized Bed Technology?', 'Pulverized Coal Technology?',
       'Stoker Technology?', 'Other Combustion Technology?',
       'Subcritical Technology?', 'Supercritical Technology?',
       'Ultrasupercritical Technology?', 'Solid Fuel Gasification System?',
       'Carbon Capture Technology?', 'Multiple Fuels?',
       'Switch Between Oil and Natural Gas?', 'Cofire Fuels?'], axis="columns")
prior_yr.head(3)
```

```python
# Sorting the updated Prior_Year dataset by the new column.

prior_yr.sort_values(by=["Plant_Name_Gen_ID"])
prior_yr.head(3)
```

```python
# Creating a new column for the Current_Year dataset.

current_yr = proposed_gen_current_yr
current_yr['Plant_Name_Gen_ID']=current_yr.apply(
    lambda x:'%s_%s' % (x["Plant Name"],x["Generator ID"]),axis=1)
current_yr.head(3)
```

```python
# Setting the new column as index for the Current_Year dataset.

current_yr.set_index("Plant_Name_Gen_ID", inplace=True)
current_yr.head(3)
```

```python
# Shrinking the Current_Year dataset by dropping coloumns to keep only relevant info.

current_yr = current_yr.drop(['Unit Code',
       'Ownership', 'Duct Burners',
       'Can Bypass Heat Recovery Steam Generator?',
       'RTO/ISO LMP Node Designation',
       'RTO/ISO Location Designation for Reporting Wholesale Sales Data to FERC',
       'Nameplate Power Factor',
       'Summer Capacity (MW)', 'Winter Capacity (MW)',
       'Associated with Combined Heat and Power System', 'Sector Name',
       'Sector', 'Previously Canceled', 'Energy Source 1', 'Energy Source 2',
       'Energy Source 3', 'Energy Source 4', 'Energy Source 5',
       'Energy Source 6', 'Turbines or Hydrokinetic Buoys',
       'Fluidized Bed Technology?', 'Pulverized Coal Technology?',
       'Stoker Technology?', 'Other Combustion Technology?',
       'Subcritical Technology?', 'Supercritical Technology?',
       'Ultrasupercritical Technology?', 'Solid Fuel Gasification System?',
       'Carbon Capture Technology?', 'Multiple Fuels?',
       'Switch Between Oil and Natural Gas?', 'Cofire Fuels?'], axis="columns")
current_yr.head(3)
```

```python
# Sorting the updated Current_Year dataset by the new column.

current_yr.sort_values(by=["Plant_Name_Gen_ID"])
current_yr.head(3)
```

```python
# Isolating the differences in the indexes

idx1 = prior_yr.index
idx2 = current_yr.index
idx1.difference(idx2)
```

```python
# Creating a new dataset Changes_YoY comprised of the diffrences between the two (2) original datasets.

Changes_YoY = current_yr.loc[idx2.difference(idx1), :]
Changes_YoY.head(3)
```

```python
# Inspecting updated dataset

Changes_YoY.describe(include=object)
```

```python
# Inspecting data types

Changes_YoY.dtypes
```

```python
# Rename Nameplate Capacity column

Changes_YoY = Changes_YoY.rename(columns={"Nameplate Capacity (MW)": "Nameplate Capacity"})
Changes_YoY.head(3)
```

```python
# Changing Nameplate Capacity type to float

Changes_YoY["Nameplate Capacity"] = Changes_YoY["Nameplate Capacity"].astype(float)
Changes_YoY.dtypes
```

```python
# Check for Missing Data

Changes_YoY.isna().sum()
```

# Analyzing the changes (or results)

In this section, we'll explore the results from comparing the two datasets, by using:
- Data Visualization
- Descriptive Statistics

## Data Visualization and Descriptive Statistics: Example(s)

```python
# Preparing data by Technology

by_Tech_count = Changes_YoY.groupby("Technology").agg({"Nameplate Capacity":"count"})
by_Tech_mean = Changes_YoY.groupby("Technology").agg({"Nameplate Capacity":"mean"})
by_Tech_sum = Changes_YoY.groupby("Technology").agg({"Nameplate Capacity":"sum"})
                                                            
print(by_Tech_count)
print(by_Tech_mean)
print(by_Tech_sum)
```

```python
# Visualizing data by Technology

fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(24, 8), sharex=True)

by_Tech_count.plot.bar(ax=ax0, title="U.S. - YoY Change in Number of Projects - by Tech", legend=False, color=("#002B54"));
by_Tech_sum.plot.bar(ax=ax1, title="U.S. - YoY Change in Est. Total Capacity Additions (in MW) - by Tech", legend=False, color=("#F3901D"));
by_Tech_mean.plot.bar(ax=ax2, title="U.S. - Avg Project Size (in MW) in YoY Change - by Tech", legend=False, color=("#63666A"));
```

```python
# Preparing data by Prime_Mover

by_Prime_Mover_count = Changes_YoY.groupby("Prime Mover").agg({"Nameplate Capacity":"count"})
by_Prime_Mover_mean = Changes_YoY.groupby("Prime Mover").agg({"Nameplate Capacity":"mean"})
by_Prime_Mover_sum = Changes_YoY.groupby("Prime Mover").agg({"Nameplate Capacity":"sum"})
                                                            
print(by_Prime_Mover_count)
print(by_Prime_Mover_mean)
print(by_Prime_Mover_sum)
```

```python
# Visualizing data by Prime_Mover

fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(24, 8), sharex=True)

by_Prime_Mover_count.plot.bar(ax=ax0, title="U.S.- YoY Change in Number of Projects - by Prime Mover", legend=False, color=("#002B54"));
by_Prime_Mover_sum.plot.bar(ax=ax1, title="U.S. - YoY Change in Est. Total Capacity Additions (in MW) - by Prime Mover", legend=False, color=("#F3901D"));
by_Prime_Mover_mean.plot.bar(ax=ax2, title="U.S. - Avg Project Size (in MW) in YoY Change - by Prime Mover", legend=False, color=("#63666A"));
```

```python
# Preparing data by State

by_State_count = Changes_YoY.groupby("State").agg({"Nameplate Capacity":"count"})
by_State_mean = Changes_YoY.groupby("State").agg({"Nameplate Capacity":"mean"})
by_State_sum = Changes_YoY.groupby("State").agg({"Nameplate Capacity":"sum"})
                                                            
print(by_State_count)
print(by_State_mean)
print(by_State_sum)
```
